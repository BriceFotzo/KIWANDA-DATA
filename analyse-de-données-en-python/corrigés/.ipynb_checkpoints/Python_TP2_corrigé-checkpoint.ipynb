# Mise en place d'un modèle de prédiction

Nous souhaitons prédire le prix des diamonds en fonction de différentes caractérisques. Le prix étant une valeur continue, il s'agit ici d'un problème de regression Etant donné les relations linéaires que nous avons mise en évidence dans partie "Analyse des variables", nous allons opter pour le modèle de régression linéaire.

```python
# récupérer les variables et la target 
x=df_diamonds.drop(columns='price')
y=df_diamonds.price
```

# Sélection de variables

Nous allons utiliser un alogrithme de dsélection de variable : RFE (Recursive Feature Elimination).
Comme son nom l'indique, le principe de cet algorithme est d'éliminer des features du modèle. En effet, nous allons entrianer les modèle à plusieurs reprises avec différents nombres de features et à chaque fois. Pour chauqe entrainement, on évaluera le score qu'on stockera dans une variable. Nous allons finalement garder le nombre de variables pour lequel on a le meilleur score.
Ensuite on déterminera quelles sont ces variables et on entrainera notre modèle final avec ces dernières.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.preprocessing import scale,StandardScaler
```

```python
# nombre de features
nof_list=np.arange(1,9)
# variable de stockage du score le plus élevé
high_score=0
# variable de stockage du nombre de feature pour ce score
nof=0
# liste des scores
score_list =[]
#standardiser les varibles 
x_ = StandardScaler().fit_transform(x)
for i in [42,0,21]:
    for test_size in  [0.10,0.15,0.2,0.25,0.3,0.35]:
        for n in range(len(nof_list)):
            X_train, X_test, y_train, y_test = train_test_split(x_,y, test_size = test_size,random_state=i)
            model = LinearRegression()
            rfe = RFE(model,nof_list[n])
            X_train_rfe = rfe.fit_transform(X_train,y_train)
            X_test_rfe = rfe.transform(X_test)
            model.fit(X_train_rfe,y_train)
            score = model.score(X_test_rfe,y_test)
            score_list.append(score)
            if(score>high_score):
                high_score = score
                nof = nof_list[n]
                test=test_size
                state=i
                
print("Le nombre optimale de features est : %d" %nof)
print("Le score avec %d features est: %f" % (nof, high_score))
print("La taille optimale de test est : %f" % (test))
print("Le random state optimale est: %d" % (state))
```

# Implémentation du modèle

Après sélection de variable, nous avos établie que notre modèle fonctionne le mieux avec 7 variables. Nous allons donc déterminer quelles sont ces variables et les utiliser par la suite.

```python
# récupérer les noms de variables
x_labels=x.columns
# styandardiser les variables
x_ = StandardScaler().fit_transform(x)
model_to_use =  LinearRegression()
# appliquer la selection des 7 variables qui nous intérressent
rfe = RFE(model_to_use, 7)
rfe = rfe.fit(x_, y)
# récupérer les ranking (1 pour les variables à conserver et n>=2 pour les autres)
ranks=rfe.ranking_
features_to_use=[label for i,label in enumerate(x_labels) if ranks[i]==1 ]
best_degree
```
